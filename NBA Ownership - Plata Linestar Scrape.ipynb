```python
# Data Scraping Initialization
from bs4 import BeautifulSoup as bs
import requests
import time

import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
# Get html text from a url
url_text = requests.get('https://www.linestarapp.com/Ownership/Sport/NBA/Site/FanDuel/PID/1308').text
# *line star has a crawl delay of 5 seconds*

#put into beautiful soup
soup = bs(url_text,'lxml')
#print(soup)
soup.prettify
```

```python
table = soup.find('tbody')
table_list = table.find_all('tr')
print(table_list)
print()
print(len(table_list))
```
